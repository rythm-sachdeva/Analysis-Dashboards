```python
import pandas as pd
import numpy as np
import plotly.express as ex

```

```python
df = pd.read_csv('test.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(columns=['Unnamed: 0','id'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean())
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Type of Travel'].value_counts() # Most People are travelling for business purpose
```

```python
df.Class.value_counts()
```

```python
ex.box(df,x='Class',y='Flight Distance',color='Class')
```

```python
ex.box(df,x='Class',y='Age',color='Class')
```

```python
ex.box(df,x='Type of Travel',y='Flight Distance',color='Type of Travel')
```

```python
ex.box(df,x='Type of Travel',y='Age',color='Type of Travel')
```

```python
travel_counts = df.groupby(["Gender", "Type of Travel"]).size().reset_index(name="count")

ex.bar(
    travel_counts,
    x="Gender",
    y="count",
    color="Type of Travel",
    barmode="group",
    title="Type of Travel by Gender"
)
```

```python
ex.scatter(df,x='Departure Delay in Minutes',y='Arrival Delay in Minutes',opacity=0.6,size='Flight Distance')
```
